```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.ndimage import gaussian_filter
import imageio as io
import matplotlib.colors as colors
import pywt
from skimage import metrics
```

```python
import requests
from PIL import Image
from io import BytesIO

img_urls = ['https://cims.nyu.edu/~cfgranda/foto.jpg']

img_array = []
for url in img_urls:
    response = requests.get(url)
    img = Image.open(BytesIO(response.content))
    img = img.convert('L')
    img = img.resize((512, 512))
    img = np.array(img)
    img_array.append(img)
```

```python
def visualize_image_and_recon(image, recon_image, figsize = (10, 5)):
    
    fig, axes = plt.subplots(nrows=1, ncols=2,  figsize = figsize)
    axes[0].imshow(image, cmap='gray')
    axes[0].set_title('Original Image')
    
    axes[1].imshow(recon_image, cmap='gray')
    axes[1].set_title(f"Reconstructed. PSNR: {metrics.peak_signal_noise_ratio(image, recon_image, data_range = 255.): .3f}")
    
    plt.show()
```

```python
image = np.array(img_array[0])
```

```python
n_levels = 3
wav_type = 'haar'
x_w = pywt.wavedec2(image, wav_type, level=n_levels)
```

```python
image.ndim
```

### Visualizing wavelet coeffs

```python
N_levels =[2,3,4,5]
x_w =[]
wav_type = 'haar'
for n_levels in N_levels:
    temp = pywt.wavedec2(image, wav_type, level=n_levels)
    x_w.append(temp)
```

```python
coeff_arrays =[]
coeff_slicess = []


for i in x_w:
    coeff_array, coeff_slices = pywt.coeffs_to_array(i)
    coeff_arrays.append(coeff_array)
    coeff_slicess.append(coeff_slices)
```

```python
perc = np.asarray([0.005, .01, 0.015, 0.02, 0.03, 0.04, 0.05, .10, .25, .50, .75])
titles=[]
coeff_n = []
coeff_slices_n =[]

for coeff_array,coeff_slices, level in zip(coeff_arrays,coeff_slicess,N_levels):
    
    for p in perc:
        coeff_n.append(np.where(coeff_array > np.quantile(coeff_array,1.0-p ), coeff_array,0))
        coeff_slices_n.append(coeff_slices)
        titles.append('percent kept:' +str(p*100)+'Number of levels:'+str(level))

coeff_f = []
reimages = []

for coeff_array,coeff_slices in zip(coeff_n, coeff_slices_n):
    coeff_correctformat = pywt.array_to_coeffs(coeff_array, coeff_slices, output_format='wavedec2')
    coeff_f.append(coeff_correctformat)

for i in coeff_f:
    recon_image = pywt.waverec2(i, wavelet=wav_type)
    reimages.append(recon_image)
    
for ri, title in zip(reimages,titles):
    visualize_image_and_recon( title,image,ri)
```

### Reconstructing the image

```python
coeff_correctformat = pywt.array_to_coeffs(coeff_array, coeff_slices, output_format='wavedec2')
```

```python
recon_image = pywt.waverec2(coeff_correctformat, wavelet=wav_type)
```

```python
visualize_image_and_recon(image, recon_image)
```
